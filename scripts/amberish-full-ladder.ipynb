```python
import itertools
run_names = [f"ai2-llm/olmo-ladder/amberish-5shot-{size}-{length}" for size, length in itertools.product(["150M", "300M", "530M", "750M", "1B"], ["1xC", "2xC", "5xC", "10xC"])]

validation_outputs = {}
downstream_outputs = {}

#run_names = run_names[:2]

for run_name in run_names:

    output_name = run_name.split("/")[-1]
    output_path = f"{WANDB_RESULTS}/{output_name}-val-all.csv"
    args = Namespace(
        wandb_names=[run_name],
        x_axis="throughput/total_tokens",
        y_axis=["eval/validation-and-bpb-and-downstream"],
        output_path=output_path,
    )
    if not os.path.exists(output_path):
        download_wb(args)
    validation_outputs[output_name] = output_path
```

### 2. Fitting a power law based on GFlops

```python
import numpy as np
import pandas as pd
import ladder
from olmo.scaling.scaling_laws.utils import validation as validation_names
```

```python
def get_gflops(run_name: str):
    run_name = run_name.split("/")[-1]
    size, length = run_name.split("-")[-2:]
    length_in_tokens = ladder.parse_length(length, ladder.parse_size(size))

    flops = ladder.MODEL_GFLOPS[size]
    return flops * length_in_tokens / 1e9
```

```python
mmlu_names = ['mmlu_stem_var',
       'mmlu_humanities_var',
       'mmlu_social_sciences_var',
       'mmlu_other_var']
```

```python
ce_loss = {}
downstream_bpb = {}
downstream_acc = {}
gflops = {}
tokens = {}
for run_name, output_path in validation_outputs.items():
    df_val = pd.read_csv(output_path)

    gflops[run_name] = get_gflops(run_name)
    tokens[run_name] = df_val["throughput/total_tokens"].values[-1]
    ce_loss[run_name] = np.mean([df_val[f"eval/{d}/CrossEntropyLoss"].values[-1] for d in validation_names])

    # Downstream bpb
    downstream_bpb[run_name] = {}
    downstream_bpb[run_name]["hellaswag_0shot"] = df_val["eval/downstream_bpb/hellaswag_rc_0shot_bpb_bpb"].values[-1]
    downstream_bpb[run_name]["hellaswag_5shot"] = df_val["eval/downstream_bpb/hellaswag_rc_5shot_bpb_bpb"].values[-1]
    downstream_bpb[run_name]["arc_easy_0shot"] = df_val["eval/downstream_bpb/arc_easy_rc_0shot_bpb_bpb"].values[-1]
    downstream_bpb[run_name]["arc_easy_5shot"] = df_val["eval/downstream_bpb/arc_easy_rc_5shot_bpb_bpb"].values[-1]
    downstream_bpb[run_name]["piqa_0shot"] = df_val["eval/downstream_bpb/piqa_rc_0shot_bpb_bpb"].values[-1]
    downstream_bpb[run_name]["mmlu_var"] = np.mean([df_val[f"eval/downstream_bpb/{d}_bpb_bpb"].values[-1] for d in mmlu_names])

    # Downstream accuracy
    downstream_acc[run_name] = {}
    downstream_acc[run_name]["hellaswag_0shot"] = df_val["eval/downstream/hellaswag_len_norm"].values[-1]
    downstream_acc[run_name]["hellaswag_5shot"] = df_val["eval/downstream/hellaswag_len_norm"].values[-1]  # TODO
    downstream_acc[run_name]["arc_easy_0shot"] = df_val["eval/downstream/arc_easy_acc"].values[-1]
    downstream_acc[run_name]["arc_easy_5shot"] = df_val["eval/downstream/arc_easy_acc"].values[-1]  # TODO
    downstream_acc[run_name]["piqa_0shot"] = df_val["eval/downstream/piqa_len_norm"].values[-1]
    downstream_acc[run_name]["mmlu_var"] = np.mean([df_val[f"eval/downstream/{d}_len_norm"].values[-1] for d in mmlu_names])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from olmo.scaling.scaling_laws.utils import get_coefficients_huber, chinchilla_flops_fit, grad_chinchilla_flops_fit

def get_x_y_coeffs_flops(validation_outputs, use_ce_loss=True, metric="hellaswag_0shot"):
    x = []
    y = []
    for run in validation_outputs:
        x.append(gflops[run])
        if use_ce_loss:
            y.append(ce_loss[run])
        else:
            y.append(downstream_bpb[run][metric])
    
    print(x, y)
    train_x = np.array(x[:-1])
    train_y = np.array(y[:-1])
    
    coefficients = get_coefficients_huber(
        train_x, train_y,
        chinchilla_flops_fit, grad_chinchilla_flops_fit,
        p0=[100, -0.2, 1.5],
        bounds=[(0, None), (None, None), (0, None)],
        disp=False,
    )
    return x, y, coefficients

def plot_fit_flops(ax, x, y, coefficients, title_suffix=str, use_ce_loss=True, metric="hellaswag_0shot"):
    train_x = np.array(x[:-1])
    np.array(y[:-1])

    ax.scatter(x, y, label=["150M", "300M", "530M", "750M", "1B"])
    ax.set_title(f"Varying flops, all {title_suffix}")
    ax.set_xlabel("GFlops (varying model size)")
    if use_ce_loss:
        ax.set_ylabel("CE Loss (val)")
    else:
        ax.set_ylabel(f"Downstream bpb ({metric})")
    x_known = np.linspace(min(train_x), max(train_x), 100)
    ax.plot(x_known, chinchilla_flops_fit(x_known, coefficients))
    x_predicted = np.linspace(max(train_x), max(x), int(100 / len(x)))
    ax.plot(x_predicted, chinchilla_flops_fit(x_predicted, coefficients), linestyle="dashed")
    
    pred_plots = chinchilla_flops_fit(x, coefficients)
    
    actual = y[-1]
    predicted = pred_plots[-1]
    error = abs(actual - predicted) * 100 / actual
    
    legends = []
    legends.append(f"Actual value = {round(actual, 3)}")
    legends.append(f"Predicted value = {round(predicted, 3)} (Error = {round(error, 3)} %)")
    
    ax.legend(legends, loc="upper right")
```

```python
fix, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4, figsize=(20, 10))

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("1xC")})
plot_fit_flops(ax1, x, y, coefficients, "1xC")

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("2xC")})
plot_fit_flops(ax2, x, y, coefficients, "2xC")

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("5xC")})
plot_fit_flops(ax3, x, y, coefficients, "5xC")

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("10xC")})
plot_fit_flops(ax4, x, y, coefficients, "10xC")

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("1xC")}, False)
plot_fit_flops(ax5, x, y, coefficients, "1xC", False)

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("2xC")}, False)
plot_fit_flops(ax6, x, y, coefficients, "2xC", False)

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("5xC")}, False)
plot_fit_flops(ax7, x, y, coefficients, "5xC", False)

x, y, coefficients = get_x_y_coeffs_flops({key: val for key, val in validation_outputs.items() if key.endswith("10xC")}, False)
plot_fit_flops(ax8, x, y, coefficients, "10xC", False)

plt.tight_layout()
plt.show()
```

```python
from scipy.optimize import curve_fit

def sigmoid(x, L ,x0, k, b):
    o = L / (1 + np.exp(-k*(x-x0))) + b
    return (o)

def get_x_y_coeffs_downstream(validation_outputs, metric="hellaswag_0shot", p0=None):
    x = []
    y = []
    for run in validation_outputs:
        x.append(1 - downstream_bpb[run][metric])
        y.append(downstream_acc[run][metric])
    
    train_x = np.array(x[:-1])
    train_y = np.array(y[:-1])
    
    p0 = p0 or [1, 0.4, 3, 0.3]
    
    coefficients, pcov = curve_fit(sigmoid, train_x, train_y, p0, maxfev=1000000)
    return x, y, coefficients

def plot_downstream(ax, x, y, coefficients, title_suffix=str, metric="hellaswag_0shot"):
    
    np.array(x[:-1])
    np.array(y[:-1])

    for x_, y_, run_ in zip(x, y, ["150m", "300m", "530m", "750m", "1b"]):
        ax.scatter(x_, y_, color="blue")
        ax.annotate(run_, (x_, y_))
    
    ax.set_title(f"bpb to task score ({title_suffix})")
    ax.set_xlabel(f"1 - downstream ({metric}) bpb")
    ax.set_ylabel(f"downstream ({metric}) len_norm")
    x = np.linspace(min(x), max(x), 100)
    ax.plot(x, sigmoid(x, *coefficients))
    
    pred_plots = sigmoid(x, *coefficients)
    
    actual = y[-1]
    predicted = pred_plots[-1]
    error = abs(actual - predicted) * 100 / actual
    
    legends = []
    legends.append(f"Actual value = {round(actual, 3)}")
    legends.append(f"Predicted value = {round(predicted, 3)} (Error = {round(error, 3)} %)")
    
    ax.legend(legends, loc="upper right")
```

```python
fix, axes = plt.subplots(4, 1, figsize=(10, 20))

the_range = ["1xC", "2xC", "5xC", "10xC"]
i = 0

for col in axes:
    print(the_range[i])
    x, y, coefficients = get_x_y_coeffs_downstream({key: val for key, val in validation_outputs.items() if key.endswith(the_range[i])})
    plot_downstream(col, x, y, coefficients, the_range[i])
    i += 1

plt.tight_layout()
plt.show()
```

### Map ce loss to downstream accuracy as a sigmoidal

```python
def get_x_y_coeffs_downstream_ce_loss(validation_outputs, metric="hellaswag_0shot"):
    x = []
    y = []
    for run in validation_outputs:
        x.append(1 - ce_loss[run])
        y.append(downstream_acc[run][metric])
    
    train_x = np.array(x[:-1])
    train_y = np.array(y[:-1])
    
    p0 = [1, 0.4, 3, 0.3]
    
    coefficients, pcov = curve_fit(sigmoid, train_x, train_y, p0, maxfev=50000)
    return x, y, coefficients

def plot_downstream_ce_loss(ax, x, y, coefficients, title_suffix=str, metric="hellaswag_0shot"):
    
    np.array(x[:-1])
    np.array(y[:-1])

    for x_, y_, run_ in zip(x, y, ["150m", "300m", "530m", "750m", "1b"]):
        ax.scatter(x_, y_, color="blue")
        ax.annotate(run_, (x_, y_))
    
    ax.set_title(f"bpb to task score ({title_suffix})")
    ax.set_xlabel("1 - ce_loss")
    ax.set_ylabel(f"downstream ({metric}) len_norm")
    x = np.linspace(min(x), max(x), 100)
    ax.plot(x, sigmoid(x, *coefficients))
    
    pred_plots = sigmoid(x, *coefficients)
    
    actual = y[-1]
    predicted = pred_plots[-1]
    error = abs(actual - predicted) * 100 / actual
    
    legends = []
    legends.append(f"Actual value = {round(actual, 3)}")
    legends.append(f"Predicted value = {round(predicted, 3)} (Error = {round(error, 3)} %)")
    
    ax.legend(legends, loc="upper right")
```

```python


fix, axes = plt.subplots(4, 1, figsize=(10, 20))

the_range = ["1xC", "2xC", "5xC", "10xC"]
i = 0

metric = "hellaswag_0shot"

#for row in axes:
for col in axes:
    print(the_range[i])
    x, y, coefficients = get_x_y_coeffs_downstream_ce_loss({key: val for key, val in validation_outputs.items() if key.endswith(the_range[i])}, metric=metric)
    plot_downstream_ce_loss(col, x, y, coefficients, the_range[i], metric=metric)
    i += 1

plt.tight_layout()
plt.show()
```
